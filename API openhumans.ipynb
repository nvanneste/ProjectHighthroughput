```python
import urllib.request

response = urllib.request.urlopen("https://www.openhumans.org/api/public-data/?source=american_gut")

AmericanGut = open("AmericanGut.json", mode='wb')
AmericanGut.write(response.read())
AmericanGut.close()

#works
```
